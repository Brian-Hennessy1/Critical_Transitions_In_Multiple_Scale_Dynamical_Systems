```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def F(t,x):
    
    y_0 = -1*x[1]
    y_1 = -1*x[0]
    return [y_0,y_1]

def G(t,x):
    return [x[1],x[0]]
```

```python
x0 = [1/2,0]
x1 = [-1/2,0]
x2 = [0,1/2]
x3 = [0,-1/2]
t = np.linspace(0,2,100)
sol1 = solve_ivp(F, [0,2],  x0,method='RK45',t_eval = t)
sol2 = solve_ivp(F, [0,2],  x1,method='RK45',t_eval = t)
sol3 = solve_ivp(F, [0,2],  x2,method='RK45',t_eval = t)
sol4 = solve_ivp(F, [0,2],  x3,method='RK45',t_eval = t)
sol5 = solve_ivp(G, [0,2],  x0,method='RK45',t_eval = t)
sol6 = solve_ivp(G, [0,2],  x1,method='RK45',t_eval = t)
sol7 = solve_ivp(G, [0,2],  x2,method='RK45',t_eval = t)
sol8 = solve_ivp(G, [0,2],  x3,method='RK45',t_eval = t)
```

```python
def arrow_numbers(p,q):
    i =30
    x_i = p[i]
    y_i = q[i]
    dx_i = p[i+1]-p[i]
    dy_i = q[i+1]-q[i]
    return x_i,y_i,dx_i,dy_i
```

```python
plt.figure(figsize = (6,6))
x = np.linspace(-1,1,9)
y = np.linspace(-1,1,9)
X,Y=np.meshgrid(x,y)

U = -Y
V = -X
#plt.quiver(X,Y,U,V,scale=8)
plt.plot(x,y,'b')
plt.plot(x,-y,'b')
plt.plot(sol1.y[0],sol1.y[1],'b')
plt.plot(sol2.y[0],sol2.y[1],'b')
plt.plot(sol3.y[0],sol3.y[1],'b')
plt.plot(sol4.y[0],sol4.y[1],'b')
plt.plot(sol5.y[0],sol5.y[1],'b')
plt.plot(sol6.y[0],sol6.y[1],'b')
plt.plot(sol7.y[0],sol7.y[1],'b')
plt.plot(sol8.y[0],sol8.y[1],'b')
plt.xlim([-1,1])
plt.ylim([-1,1])
arrow_sol1 = arrow_numbers(sol1.y[0],sol1.y[1])
arrow_sol2 = arrow_numbers(sol2.y[0],sol2.y[1])
arrow_sol3 = arrow_numbers(sol3.y[0],sol3.y[1])
arrow_sol4 = arrow_numbers(sol4.y[0],sol4.y[1])
arrow_sol5 = arrow_numbers(sol5.y[0],sol5.y[1])
arrow_sol6 = arrow_numbers(sol6.y[0],sol6.y[1])
arrow_sol7 = arrow_numbers(sol7.y[0],sol7.y[1])
arrow_sol8 = arrow_numbers(sol8.y[0],sol8.y[1])
plt.arrow(arrow_sol1[0],arrow_sol1[1],-arrow_sol1[2],-arrow_sol1[3],width=0.04)
plt.arrow(arrow_sol2[0],arrow_sol2[1],-arrow_sol2[2],-arrow_sol2[3],width=0.04)
plt.arrow(arrow_sol3[0],arrow_sol3[1],-arrow_sol3[2],-arrow_sol3[3],width=0.04)
plt.arrow(arrow_sol4[0],arrow_sol4[1],-arrow_sol4[2],-arrow_sol4[3],width=0.04)
plt.arrow(arrow_sol5[0],arrow_sol5[1],arrow_sol5[2],arrow_sol5[3],width=0.04)
plt.arrow(arrow_sol6[0],arrow_sol6[1],arrow_sol6[2],arrow_sol6[3],width=0.04)
plt.arrow(arrow_sol7[0],arrow_sol7[1],arrow_sol7[2],arrow_sol7[3],width=0.04)
plt.arrow(arrow_sol8[0],arrow_sol8[1],arrow_sol8[2],arrow_sol8[3],width=0.04)
plt.arrow(1/2,-1/2,-1/10,1/10,width=0.04)
plt.arrow(-1/2,-1/2,-1/10,-1/10,width=0.04)
plt.arrow(1/2,1/2,1/10,1/10,width=0.04)
plt.arrow(-1/2,1/2,1/10,-1/10,width=0.04)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Normal form of a Saddle Point.")
plt.scatter(0,0,color='k',s = 400,label="Fixed Point")
plt.legend()
plt.savefig("Normal_Form_explanation")
```
