```python
# Import packages

import numpy as np
from matplotlib import pyplot as plt
from scipy.integrate import odeint
%matplotlib inline

# Define vector field
k = 0.01
nu = 0.8
phi = 0.12


#specify variable values



def vField(x, t):
    #u = a/(1+b_1*x[0]+b_2*x[1])*(x[0])-c*x[0]
    #v = A/(1+B*(x[0]+x[1]))*(x[1])-C*x[1]
    u = x[0]*(1-phi*x[0])-(x[0]*x[1])/(1+nu*x[0])
    v = k*((x[0]*x[1])/(1+nu*x[0])-x[1])
    return [u, v]
    
    
# Plot vector field

#X, Y = np.mgrid[-np.pi:np.pi:10j,-7:7:10j]
X, Y = np.mgrid[1:2.5:15j,0:2:15j]
U, V = vField([X,Y],0)

fig, ax = plt.subplots(figsize=(10, 7))
#ax.quiver(X, Y, U, V)

# Compute and plot 4 trajectories
'''
t=np.linspace(0,4000,2000)
x0, y0 = 0.2,1.5
x = odeint(vField,[x0,y0],t)
ax.plot(x[:,0],x[:,1]);

t=np.linspace(0,4000,2000)
x0, y0 = 0.01,1.01
x = odeint(vField,[x0,y0],t)
ax.plot(x[:,0],x[:,1]);
'''



#plt.plot(0, 80000, 'ro', label = '$(0, D)$')
#plt.plot(20000, 0, 'bo', label = '$(d, 0)$')
#for i in range(1, 5):
 #   plt.plot(0, i-0.5, 'cx', label = '$u=0$')


#plot quadratic 
L = [] 
for i in np.linspace(0, 9, 200):
    L.append(1+(nu-phi)*i-((phi*nu)*(i**2)))
    
plt.plot(np.linspace(0, 9, 200), L, 'r--', label = '$C_0^1$', linewidth = 5)


M = [] 
for i in np.linspace(0, 3.6, 200):
    M.append(1+(nu-phi)*i-((phi*nu)*(i**2)))
    
plt.plot(np.linspace(0, 3.6, 200), M, 'b--', linewidth = 5)



plt.axvline(x=0, color = 'r', label = '$C_0^0$', linewidth = 5)
plt.axvline(x=0, ymin = 0, ymax = 0.58, color = 'b', linewidth = 5)

#plt.plot(0, 1, 'mo', label = '$(0, 1)$')
plt.plot(0, 0, 'co', label = '$e_1$', markersize=18)
plt.plot(1/phi,0, 'mo', label = '$e_2$',markersize=18)
plt.plot(1/(1-nu), (1-nu-phi)/((1-nu)**2), 'yo', label = '$e_3$', markersize=15)
plt.plot((nu-phi)/(2*nu*phi), (nu+phi)**2/(4*nu*phi), 'go', label = 'Fold', markersize=18)
plt.legend(fontsize=18, loc = "upper right")
#plt.title(label = '$(b_1/b_2)d<D$ - (Accelerated-acute Phase)')

plt.xlabel('x', fontsize = 40)
plt.ylabel('y', rotation = 0, fontsize = 40)
plt.xticks(size = 30)
plt.yticks(size = 30)
plt.xlim([-0.5, 9])
plt.ylim([-0.05, 2.3])
#plt.savefig('S_0_case1.png')
plt.show()
```
